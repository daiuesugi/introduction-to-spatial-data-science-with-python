```python
import numpy as np
import pandas as pd
from sklearn.datasets import fetch_california_housing

from Functions.ols import OLS  # 自作モジュールの読み込み
```

```python
# カリフォルニア住宅価格データセットの読み込み
california_data = fetch_california_housing()
```

```python
# データフレームの作成
df_X = pd.DataFrame(california_data.data, columns=california_data.feature_names)
df_y = pd.DataFrame(california_data.target, columns=california_data.target_names)
```

```python
# データフレームをnp.arrayに変換する
data = df_X.loc[:, ['MedInc', 'HouseAge', 'AveRooms', 'AveBedrms', 'Population', 'AveOccup']].values
# 定数項の列(すべて1)を作成する
intercept = np.ones(df_X.shape[0]).reshape([-1, 1])

# 2つの配列を水平方向に結合する
X = np.hstack([intercept, data])
y = df_y.values.reshape([-1, 1])
```

```python
# OLSオブジェクトの作成
ols = OLS(y, X)
```

```python
# オブジェクト型の確認
type(ols)
```

```python
# OLS推定の実行
ols.fit()
```

```python
# 係数パラメータ推定値
ols.beta_hat
```

```python
# 推定結果の格納
summary = ols.summary()
print(summary)
```
