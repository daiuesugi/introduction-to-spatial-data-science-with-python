```python
import numpy as np
import geopandas as gpd
import matplotlib.pyplot as plt

import libpysal as ps  # サンプルデータセット"georgia"をインポートするために必要なモジュール

from Functions.gwr import GWR
from Functions.sel_bw import Sel_BW
from Functions.utils import *
```

```python
# サンプルデータ"georgia"をインポートしてgeopandasデータフレームを作成
georgia = gpd.read_file(ps.examples.get_path('G_utm.shp'))
```

```python
y, X, coords = get_input_data(georgia['PctBach'], georgia[['PctFB', 'PctBlack', 'PctRural']], georgia[['X', 'Y']], std=False)
```

```python
selector = Sel_BW(y, X, coords)
```

```python
bw = selector.search()
print(bw)
```

```python
model = GWR(y, X, coords, bw)
```

```python
model.fit()
print(model.beta_hat)
```

```python
# 推定結果の表示
print(model.summary())
```
